```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Thresholds for decision making
TEMP_THRESHOLD = 35       # Â°C
HUMIDITY_THRESHOLD = 40   # %
SOIL_MOISTURE_THRESHOLD = 30  # %

# Lists to store data
readings = []
temperature_list = []
humidity_list = []
moisture_list = []
pump_status_list = []

# Generate simulated data
for i in range(20):  # 20 sample readings
    temperature = round(random.uniform(25, 45), 2)
    humidity = round(random.uniform(20, 60), 2)
    soil_moisture = round(random.uniform(10, 50), 2)

    # Decision logic for pump
    if soil_moisture < SOIL_MOISTURE_THRESHOLD and temperature > TEMP_THRESHOLD and humidity < HUMIDITY_THRESHOLD:
        pump_status = "ON"
    else:
        pump_status = "OFF"

    # Append data
    readings.append(i+1)
    temperature_list.append(temperature)
    humidity_list.append(humidity)
    moisture_list.append(soil_moisture)
    pump_status_list.append(pump_status)

# Create DataFrame
df = pd.DataFrame({
    "Reading": readings,
    "Temperature (Â°C)": temperature_list,
    "Humidity (%)": humidity_list,
    "Soil Moisture (%)": moisture_list,
    "Pump Status": pump_status_list
})

# Display table
print(df)

# Plotting the data
plt.figure(figsize=(12, 6))

plt.plot(df["Reading"], df["Temperature (Â°C)"], label='Temperature (Â°C)', color='red', marker='o')
plt.plot(df["Reading"], df["Humidity (%)"], label='Humidity (%)', color='blue', marker='x')
plt.plot(df["Reading"], df["Soil Moisture (%)"], label='Soil Moisture (%)', color='green', marker='s')

plt.axhline(y=TEMP_THRESHOLD, color='red', linestyle='--', label='Temp Threshold')
plt.axhline(y=HUMIDITY_THRESHOLD, color='blue', linestyle='--', label='Humidity Threshold')
plt.axhline(y=SOIL_MOISTURE_THRESHOLD, color='green', linestyle='--', label='Moisture Threshold')

plt.title("Smart Irrigation System - Sensor Readings Simulation")
plt.xlabel("Reading Number")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
